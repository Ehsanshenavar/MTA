```python
import pandas as pd
import plotly.graph_objects as go

# read CSV
df = pd.read_csv('./top stations.csv')

# Cumulative Percentage
df['Cumulative Percentage'] = df['18814126'].cumsum() / df['18814126'].sum() * 100

# Pareto chart
fig = go.Figure()

# bar charts
fig.add_trace(go.Bar(
    x=df['1'],
    y=df['18814126'],
    name='Ridership'
))

# Cumulative path
fig.add_trace(go.Scatter(
    x=df['1'],
    y=df['Cumulative Percentage'],
    name='Cumulative Percentage',
    yaxis='y2'
))

# chart setting
fig.update_layout(
    title='Pareto Chart of Top 10 Stations by Ridership',
    xaxis=dict(title='Station'),
    yaxis=dict(title='Ridership'),
    yaxis2=dict(title='Cumulative Percentage', overlaying='y', side='right'),
    showlegend=False
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the 'transit_timestamp' to datetime format
df['transit_timestamp'] = pd.to_datetime(df['transit_timestamp'])

# Extract the hour from the timestamp
df['hour'] = df['transit_timestamp'].dt.hour

# Group by hour and calculate the mean ridership
hourly_ridership = df.groupby('hour')['ridership'].mean().reset_index()

# Plot the mean ridership by hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_ridership, x='hour', y='ridership', marker='o')
plt.title('Average Ridership by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Ridership')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Extract the day from the timestamp
df['day'] = df['transit_timestamp'].dt.day

# Group by day and calculate the mean ridership
daily_ridership = df.groupby('day')['ridership'].mean().reset_index()

# Plot the mean ridership by day
plt.figure(figsize=(31, 6))
sns.lineplot(data=daily_ridership, x='day', y='ridership', marker='o')
plt.title('Average Ridership by Day of Month')
plt.xlabel('Day of Month')
plt.ylabel('Average Ridership')
plt.grid(True)
plt.show()
```
